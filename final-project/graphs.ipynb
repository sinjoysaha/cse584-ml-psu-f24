```python
import matplotlib.pyplot as plt
from collections import Counter
from typing import List, Dict
```

```python
def analyze_and_plot_categories(items: List[List[str]], title="", xlab="Laws or Constraints", color="teal"):
    # Flatten the list of categories
    all_categories = [category for sublist in items for category in sublist]
    
    # Count frequency of each category
    category_counts = Counter(all_categories)
    total_items = len(items)
    
    # Calculate percentages
    category_percentages = {k: (v / total_items) * 100 for k, v in category_counts.items()}
    
    # Prepare data for plotting
    categories = list(category_counts.keys())
    counts = list(category_counts.values())
    percentages = [category_percentages[c] for c in categories]
    print(counts)
    print(percentages)
    
    # Plot data
    fig, ax1 = plt.subplots(figsize=(10, 6))
    
    # Bar plot for counts
    ax1.bar(categories, percentages, color=color)
    ax1.set_xlabel(xlabel=xlab)
    ax1.set_ylabel('% of Questions')
    ax1.tick_params(axis='y')
    ax1.set_title(title)
    ax1.grid(True, axis='y')
    plt.xticks(rotation=45, ha='right')
    
    # # Line plot for percentages on a secondary y-axis
    # ax2 = ax1.twinx()
    # ax2.plot(categories, percentages, label='Percentages')
    # ax2.set_ylabel('Percentages (%)', color='orange')
    # ax2.tick_params(axis='y', labelcolor='orange')
    
    # Add legends
    fig.tight_layout()
    # plt.legend(loc='upper left')
    plt.show()
    
    # Print summary table
    print("Category Analysis:")
    print(f"{'Category':<20} {'Count':<10} {'Percentage (%)':<15}")
    print("-" * 45)
    for category in categories:
        print(f"{category:<20} {category_counts[category]:<10} {category_percentages[category]:<15.2f}")
    
```

```python
import pandas as pd
import numpy as np
```

```python
def plot_metrics(df, runs, xlab, title):
    bar_width = 0.1  # Width of the bars
    x = np.arange(len(runs))

    fig, ax = plt.subplots(figsize=(8, 5))
    for i, metric in enumerate(["accuracy", "f1", "precision", "recall"]):
        ax.bar(x - ((2-i)*bar_width) , df[f"test_{metric}"], bar_width, label=metric)

    ax.set_xlabel(xlab)
    ax.set_ylabel('Values')
    ax.set_title(title)
    ax.set_xticks(x)
    ax.set_xticklabels(runs)
    ax.legend()
    ax.grid(True)
    plt.show()
```
